```julia
using LightGraphs
```

```julia
# Creating an empty graph
G = Graph()
```

```julia
# Create a 5 node graph with 10 random edges
G = Graph(5, 10)
```

```julia
# Add an edge between nodes 3 and 5
add_edge!(G, 3, 5)
```

```julia
# Add a 6th node to G
add_vertex!(G)
```

```julia
# Return the number of nodes
nv(G)
```

```julia
# Return the number of edges
ne(G)
```

```julia
# Retrun the neigbors of the new node 6
neighbors(G, 6)
```

```julia
# Create a function that checks if there are edges between all pairs
# of nodes
function identify_edges(g)
    for v1 in 1:nv(g)
        for v2 in 1:nv(g)
            println(v1, "\t", v2, "\t", has_edge(g, v1, v2))
        end
    end
end
```

```julia
identify_edges(G)
```

```julia
function find_all_edges(g)
    println("Node", "\t", "All edges to node")
    for n in 1:nv(g)
        println("v", n , "\t", in_edges(g, n))
    end
end
```

```julia
find_all_edges(G)
```

```julia
# Degree of all nodes
function degree_all_nodes(g)
    println("Node", "\t", "Degree")
    for n in 1:nv(g)
        println("n", n, "\t", degree(g, n))
    end
end
```

```julia
degree_all_nodes(G)
```

```julia
# All neighbors
function all_neighbors(g)
    println("Node", "\t", "Neighbors")
    for n in 1:nv(g)
        println("n", n, "\t", neighbors(g, n))
    end
end
```

```julia
all_neighbors(G)
```
